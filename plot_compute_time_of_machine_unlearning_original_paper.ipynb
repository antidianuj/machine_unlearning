```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# create a meshgrid
x = np.linspace(1, 5, 100)
y = np.linspace(1, 10, 100)
S,K = np.meshgrid(x, y)


# define the function to plot
def f(S,K,N):
    return (N/S+1/(2*S)-1)*K-K**2/(2*S)

Z=f(S, K, 10000)
# Z2=f(D,R, 100)

# create the 3D plot
fig = plt.figure(figsize=(4,4),dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(S,K, Z, cmap='viridis')   #epoch is kept constant at 10
# ax.plot_surface(D, R, Z2, cmap='coolwarm')
ax.set_xlabel('# shards')
ax.set_ylabel('# Unlearning Requests')
ax.set_zlabel('# compute time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# create a meshgrid
x = np.linspace(1, 5, 100)
y = np.linspace(1, 10, 100)
R,D = np.meshgrid(x, y)


# define the function to plot
def f(D, R, ep):
    return ep*D*(2/3+1/3*R)

Z=f(D, R, 100)
Z2=f(D,R, 100)

# create the 3D plot
fig = plt.figure(figsize=(4,4),dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(D,R, Z, cmap='viridis')   #epoch is kept constant at 10
# ax.plot_surface(D, R, Z2, cmap='coolwarm')
ax.set_xlabel('# points per shard')
ax.set_ylabel('# slices')
ax.set_zlabel('# compute time')
plt.show()
```
